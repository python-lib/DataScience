```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = x * 2
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)

axes[0][0].set_title('AA')
axes[0][1].set_title('AB')
axes[1][0].set_title('BA')
axes[1][1].set_title('BB')

fig.suptitle('Figures Level')

# for ax in axes:
#     ax.plot(x,y)


plt.tight_layout()
```
