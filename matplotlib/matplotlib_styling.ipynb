```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label="X vs X", color='#1d05f5', lw=10)
lines = ax.plot(x,x**2,label="X vs X^2", color='#18c729', lw=2)
lines[0].set_dashes([1,1,1,2,3,5])
#ax.legend(loc=4)
ax.legend(loc=(1.1,0.5))
```
