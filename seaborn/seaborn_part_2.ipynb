```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.displot(data=df, x='salary')
sns.set(style='dark')
```

```python
sns.histplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages, x='age')
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=1, fill=True)
```

# CATEGORICAL PLOTS

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, errorbar='sd', hue='division')
plt.legend(bbox_to_anchor=(1.05, 1))
```

whiskers are defined by 1.5 x IQR
outside the whiskers are outliers
