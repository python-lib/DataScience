```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxenplot(data=df, y='reading score', x='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, y='reading score', x='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
sns.swarmplot(data=df, x='math score', size=2)
```
